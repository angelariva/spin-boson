# Spin-Boson simulations' results, $\alpha = 0.01$, $\alpha = 0.05$, $\alpha = 0.3$
_________________________________________

### Simulation parameters
The Hamiltonian of the system is the spin-boson Hamiltonian defined by:
$$
H =  \frac{\epsilon}{2}\sigma_z +  \frac{\sigma_x}{2} \otimes \sum_k g_k(b_k^\dagger+b_k) + \sum_k \omega_k b_k^\dagger b_k, 
$$
with $\epsilon=0.2$. The spectral density function $J$ is defined as:
$$
J(\omega) = 2 \alpha \omega \Big(\frac{\omega}{\omega_c}\Big)^{s-1}\theta(\omega - \omega_c),
$$
where the cutoff frequency is set to $\omega_c=1$. For different values of $s$, the value of $\alpha$ has been rescaled in the following way: 
$$\alpha(s) = \frac{\alpha}{\omega_0^s}$$

There exists a unitary transformation that defines a chain mapping of this Hamiltonian from a star configuation to a chain configuration. Such a mapping is characterized by a set of coefficients, that are computed using standard integration techniques. Using the T-TEDOPA chain mapping it is possible to perform simulations at finite temperature, where the temperature effects are absorbed in the thermal spectral density:
$$
J(\omega, \beta) = \text{sign}(\omega) \frac{J(|\omega|)}{2} \Big( 1+ \coth\Big(\frac{\omega \beta}{2}\Big)\Big).
$$

NB: the rescaled $\alpha(s)$ affects the chain mapping, that has to be modified accordingly.

The simulations of the dynamics are performed using MPS methods, and in particular the `DTDVP` algorithm. 

### Data analysis parameters
To fit the evolution in time of the mean values of the Pauli operators, the following function was used:  
$$
a e^\Gamma + c.
$$
The expected relation for relaxation time $T_1$, dephasing time $T_2$ and pure dephasing time $T^*$ is:  
$$
\frac{1}{T_2} = \frac{1}{2 T_1} + \frac{1}{T^*}
$$

# Populations and coherences

```python
def func(x, a, b, c):
    return a*np.exp(-b * x)+c 

def funcx(x, a, b, c):
    return np.cos(a*x)*np.exp(-b * x)+c


gammaz = np.zeros((len(beta)))
gammax = np.zeros((len(beta)))


def spin(s, save):
    f = pl.figure(figsize=(25,15))
    t = sz_res["2.0/1.0/times"][()]
    step = 1
    gammaz = []
    gammax = []
    for b in beta:
        pl.subplot(2,2,step)
        spz = sz_res[b+"/"+s+"/sz"][()]
        spx = sx_res[b+"/"+s+"/sx"][()]
        pl.plot(t, spz, label=r'$ \langle \sigma_z \rangle $') 
        pl.plot(t, spx, label=r'$ \langle \sigma_x \rangle $') 
        poptz, pcovz = curve_fit(func, t, spz, maxfev=3000)
        poptx, pcovx = curve_fit(funcx, t, spx, maxfev=300000)
        pl.plot(t, func(t, *poptz), label='fit: $a_z$=%5.3f, $\Gamma_z$=%5.3f, $c_z$=%5.3f' % tuple(poptz))
        pl.plot(t, funcx(t, *poptx), label='fit: $\omega_x$=%5.3f, $\Gamma_x$=%5.3f, $c_x$=%5.3f' % tuple(poptx))
 #       pl.plot(t, func(t, *poptz), label='fit: $a_z$=%5.3f, $\Gamma_z$=%5.3f' % tuple(poptz))
 #       pl.plot(t, funcx(t, *poptx), label='fit: $\omega_x$=%5.3f, $\Gamma_x$=%5.3f' % tuple(poptx))

        gammaz.append(poptz[1])
        gammax.append(poptx[1])
        pl.xlabel(r'$\omega_c t$', fontsize=16)
        pl.ylabel(r'$\langle \sigma \rangle$', fontsize=16)
        pl.title(r'$\beta = $' + b, fontsize=18)
        pl.legend(fontsize=18)
        pl.grid()
        step+=1
    pl.suptitle(r"$s = $"+s, fontsize=20, x=0.05, y=0.52)
    if save: f.savefig("spin_s="+s+".pdf", bbox_inches='tight')
    pl.show()
    return (gammaz, gammax)

def mini_spin(coup, ohm, b, save):
    f = pl.figure(figsize=(10,5)) 
    if coup=="weak":
        t = sz_res["2.0/1.0/times"][()]
        spz = sz_res[b+"/"+ohm+"/sz"][()]
        spx = sx_res[b+"/"+ohm+"/sx"][()]
    elif coup=="strong":
        t = sz_strong["2.0/1.0/times"][()]
        spz = sz_strong[b+"/"+ohm+"/sz"][()]
        spx = sx_strong[b+"/"+ohm+"/sx"][()]
    else:
        t = sz_s3["2.0/3.0/times"][()]
        spz = sz_s3[b+"/3.0/sz"][()]
        spx = sx_s3[b+"/3.0/sx"][()]
    pl.plot(t, spz, label=r'$ \langle \sigma_z \rangle $') 
    pl.plot(t, spx, label=r'$ \langle \sigma_x \rangle $') 
    poptz, pcovz = curve_fit(func, t, spz, maxfev=3000)
    poptx, pcovx = curve_fit(funcx, t, spx, maxfev=300000000)
    pl.plot(t, func(t, *poptz), label='fit: $a_z$=%5.3f, $\Gamma_z$=%5.3f, $c_z$=%5.3f' % tuple(poptz))
    pl.plot(t, funcx(t, *poptx), label='fit: $\omega_x$=%5.3f, $\Gamma_x$=%5.3f, $c_x$=%5.3f' % tuple(poptx))
#    pl.plot(t, func(t, *poptz), label='fit: $a_z$=%5.3f, $\Gamma_z$=%5.3f' % tuple(poptz))
#    pl.plot(t, funcx(t, *poptx), label='fit: $\omega_x$=%5.3f, $\Gamma_x$=%5.3f' % tuple(poptx))
    pl.xlabel(r'$\omega_c t$', fontsize=16)
    pl.ylabel(r'$\langle \sigma \rangle$', fontsize=16)
    pl.legend(fontsize=15)
    #pl.title(tit, x=0.05)
    pl.grid()
    if coup=="weak":
        if save: f.savefig("spin_s="+ohm+"_beta="+b+".pdf", bbox_inches='tight')
        else: pl.suptitle(r"$s = $"+ohm+r"$,\quad \beta = $"+b+", weak", fontsize=20)
    elif coup=="strong":
        if save: f.savefig("spin_strong_s="+ohm+"_beta="+b+".pdf", bbox_inches='tight')
        else: pl.suptitle(r"$s = $"+ohm+r"$,\quad \beta = $"+b+", strong", fontsize=20)
    else: 
        if save: f.savefig("spin_s3_s="+ohm+"_beta="+b+".pdf", bbox_inches='tight')
        else: pl.suptitle(r"$s = $"+ohm+r"$,\quad \beta = $"+b+", s3", fontsize=20)
    return (poptz[1], poptx[1])
```

```python
def rates(gam):
    pl.figure(figsize=(16,5))
    marks = ["d","p","*","<","3","o","v","2",">","1"]
    
    pl.subplot(1,2,1)
    step = 0
    for ohm in s:
        stringa="-"+marks[step]
        pl.plot(beta, gam[step][0], stringa, label=r'$s = $' +ohm, markersize=10)
        step += 1
    pl.legend(loc="best")
    pl.xlabel(r'$\beta$', fontsize=16)
    pl.ylabel(r'$\Gamma_z$', fontsize=16)
    pl.title("Relaxation rate"+", " + r"$\quad |\psi_0 \rangle = |1 \rangle$ ")
    pl.subplot(1,2,2)
    step = 0    
    for ohm in s:
        stringa="-"+marks[step]
        pl.plot(beta, gam[step][1], stringa, label=r'$s = $' +ohm, markersize=10)
        step += 1
    pl.legend(loc="best")
    pl.xlabel(r'$\beta$', fontsize=16)
    pl.ylabel(r'$\Gamma_x$', fontsize=16)
    pl.title("Dephasing rate"+", " + r"$ \quad | \psi_0 \rangle = \frac{1}{2}(|1 \rangle + |0 \rangle)$ ")
    pl.show()
    
    
def rates_new(gam):
    pl.figure(figsize=(16,5))
    marks = ["d","p","*","<","3","o","v","2",">","1"]
    
    pl.subplot(1,2,1)
    step = 0
    for b in beta:
        stringa="-"+marks[step]
        gs = [Gamma[i][0][step] for i in range(len(s))]
        pl.plot(s, gs, stringa, label=r'$\beta = $' +b, markersize=10)
        step += 1
    pl.legend(loc="best")
    pl.xlabel(r'$s$', fontsize=16)
    pl.ylabel(r'$\Gamma_z$', fontsize=16)
    pl.title("Relaxation rate"+", " + r"$\quad |\psi_0 \rangle = |1 \rangle$ ")
    pl.subplot(1,2,2)
    step = 0    
    for b in beta:
        gs = [Gamma[i][1][step] for i in range(len(s))]
        stringa="-"+marks[step]
        pl.plot(s, gs, stringa, label=r'$\beta = $' +b, markersize=10)
        step += 1
    pl.legend(loc="best")
    pl.xlabel(r'$s$', fontsize=16)
    pl.ylabel(r'$\Gamma_x$', fontsize=16)
    pl.title("Dephasing rate"+", " + r"$ \quad | \psi_0 \rangle = \frac{1}{2}(|1 \rangle + |0 \rangle)$ ")
    pl.show()
```

```python
def comparison(gam, s):
    f = pl.figure(figsize=(21,3))
    marks = ["d","p","*","<","3","o","v","2",">","1"]
    step = 1    
    for ohm in range(len(s)):
        pl.subplot(1,len(s),step)
        if step==1: pl.ylabel(r'$\Gamma$', fontsize=16)
        gf = [gam[ohm][1][i]-0.5*gam[ohm][0][i] for i in range(len(gam[ohm][0]))]
        pl.plot(beta, gam[ohm][0], '-d', label=r'$\Gamma_z $', markersize=10)
        pl.plot(beta, gam[ohm][1], '-p', label=r'$\Gamma_x $', markersize=10)
        pl.plot(beta, gf, '-*', label=r'$\Gamma_\phi $', markersize=10)

        step += 1
        pl.legend(loc="best")
        pl.xlabel(r'$\beta$', fontsize=16)
        pl.title("$s = $"+ s[ohm])
        pl.grid()
        f.savefig("times.pdf", bbox_inches='tight')
    pl.show()
    
def comparetimes(gam):
    pl.figure(figsize=(25,3.))
    marks = ["d","p","*","<","3","o","v","2",">","1"]
    step = 1    
    for ohm in range(len(s)):
        pl.subplot(1,len(s),step)
        if step==1: pl.ylabel(r'$T$', fontsize=16)
        t1=[1/el for el in gam[ohm][0]]
        t2=[1/el for el in gam[ohm][1]]
      #  tf=[1/(1/t2[i]-1/(2*t1[i])) for i in range(len(t1))]
        pl.plot(beta, t1, '-d', label=r'$T_1 $', markersize=10)
        pl.plot(beta, t2, '-p', label=r'$T_2 $', markersize=10)
      #  pl.plot(beta, tf, '-*', label=r'$T_\phi $', markersize=10)
        step += 1
        pl.legend(loc="best")
        pl.xlabel(r'$\beta$', fontsize=16)
        pl.title("$s = $"+ s[ohm])
        pl.grid()
    pl.show()
```

```python
def single_comparison(gam, s):
    pl.figure(figsize=(5, 3))
    marks = ["d","p","*","<","3","o","v","2",">","1"]
    pl.ylabel(r'$\Gamma$', fontsize=16)
    gf = [gam[i][1]-0.5*gam[i][0] for i in range(len(gam))]
    gz = [gam[i][0] for i in range(len(gam))]
    gx = [gam[i][1] for i in range(len(gam))]
    pl.plot(beta, gz, '-d', label=r'$\Gamma_z $', markersize=10)
    pl.plot(beta, gx, '-p', label=r'$\Gamma_x $', markersize=10)
    pl.plot(beta, gf, '-*', label=r'$\Gamma_\phi $', markersize=10)
    pl.legend(loc="best")
    pl.xlabel(r'$\beta$', fontsize=16)
    pl.title(r"$s = $"+ s, fontsize=18)
    pl.grid()
    pl.show()
```
